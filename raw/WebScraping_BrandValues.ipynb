```python
# only run if you have installed Selenium on your local computer!
from bs4 import BeautifulSoup

import requests

import numpy as np
import pandas as pd

# if above solutions did not work or you ran into problems with the PATH variable, try the following: driver = selenium.webdriver.Chrome(chrome_path = <PATH_TO_FILE>)
# tip: if your Chromedriver chrashes or does not work as expected throughout the exercises, run this cell again to create a new instance of the chromedriver
```

```python
# 2007
del df

url = "https://www.rankingthebrands.com/The-Brand-Rankings.aspx?rankingID=37&year=37"

# make a get request to the URL
request_object = requests.get(url)

# return the source code from the request object
source_code = request_object.text

soup = BeautifulSoup(source_code, "html.parser")

name_data = soup.find_all(class_="name")

name_list = []


for name in name_data:
    name = name.text
    
    name_list.append(name)
    
   
 
    # Brand Rankings : 
rank_data = soup.find_all(class_="pos")

rank_list = []

for i in rank_data:
    rank = i.text
    
    rank_list.append(rank)
        
     # Brand Value : 
        
brandvalue_data = soup.find_all(class_="weighted")    
    
brandvalue_list = []

for i in brandvalue_data:
    brandvalue = i.text
    
    brandvalue_list.append(brandvalue)
               
    # Year :
        
year = np.ones(100)*2007
    
    # Creating the Data Frame and Excel File :
    
df = pd.DataFrame(list(zip(year, rank_list[1:], name_list[1:], brandvalue_list)), columns=["year","Brand Rank", 'Brand Names', "Brand Value"])


df.to_excel('BrandNames2007.xlsx', index=False)

```
